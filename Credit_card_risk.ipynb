```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

This dataset is highly unbalanced


0--> Normal Transanction


1--> Fraudlant transanction

```python
#Separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#Statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both the transaction
credit_card_data.groupby('Class').mean()
```

Under-Sampling



Build a sample dataset containing similar distribution of normal transanction and Fraudulent Transanction



number of Fraudlent transanction is 492

```python
legit_sample = legit.sample(n = 492)
```

concatenating two dataframe

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the Data into features and Targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training


Logistic Refression


```python
model = LogisticRegression()
```

```python
#training the logistic regression with Training data 
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data: ', test_data_accuracy)
```
