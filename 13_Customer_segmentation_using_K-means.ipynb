```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# loading the data from csv file to pandas data frame
customer_data = pd.read_csv('/content/drive/MyDrive/Datasets /Mall_Customers.csv')
```

Choosing the Annual Income Column & Spending Score column

```python
X = customer_data.iloc[:,[3,4]].values
```

```python
print(X)
```

Choosing the number of clusters


WCSS- Within clusters Sum of square

```python
# Finding wcss for different number of clusters
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
# Plot an Elbow graph
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)

# return a label for each data point based on their cluster
Y = kmeans.fit_predict(X)
print(Y)
```

```python
# Plotting all the clusters and their centroid

plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0], X[Y==0,1], s = 25, c = 'green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s = 25, c = 'red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s = 25, c = 'yellow', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s = 25, c = 'orange', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s = 25, c = 'blue', label='Cluster 5')

# plot all the clusters and centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='centroids')
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
